```python
# Import the Libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.metrics import log_loss, ConfusionMatrixDisplay, average_precision_score, accuracy_score
from sklearn.metrics import roc_curve, precision_recall_curve, auc, f1_score, confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r"E:\wisc_bc_ContinuousVar.csv")
df.head()
```

```python
print(df.shape)
```

```python
# to describe the data info
summary = df.describe()
print(summary)
```

```python
df.isnull().sum()
```

```python
split_1 = df.drop(columns = ['id','diagnosis'])
split_2 = df['diagnosis']
```

```python
split_1
```

```python
split_2 
```

```python
split_1_scaled = StandardScaler()
split_2_scaled = StandardScaler()
split_1 = split_1_scaled.fit_transform(split_1)
```

```python
split_2 = split_2.astype("category")
split_2 = split_2.replace({'B': 0, 'M': 1})
```

```python
split_1
```

```python
split_2
```

```python
#splitting the data into training and testing datasets.
split_1_train, split_1_test,split_2_train, split_2_test = train_test_split(split_1, split_2 , test_size=0.3)
```

```python
svm_model = svm.SVC(kernel='linear') # Linear Kernel
svm_model.fit(split_1_train, split_2_train)
y_prediction = svm_model.predict(split_1_test)
```

```python
accuracy = accuracy_score(split_2_test, y_prediction)
print('Accuracy:', accuracy*100, "%")
```

```python
results = confusion_matrix(split_2_test, y_prediction)
print(results)
print('Classification Report')
print(classification_report(split_2_test, y_prediction))
```
