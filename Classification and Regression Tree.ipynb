```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import log_loss, ConfusionMatrixDisplay, average_precision_score, accuracy_score
from sklearn.metrics import roc_curve, precision_recall_curve, auc, f1_score, confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#load the data
tr = pd.read_csv(r"E:\breast-cancer-wisconsin.csv")
tr =pd.DataFrame(tr)
tr.head()
```

```python
# Summarizing each column
getting_summary = tr.describe()
print(getting_summary)
```

```python
tr= tr[tr != '?'].dropna()
print(tr[tr.isna().any(axis=1)])

split_1 = tr.drop(columns = ['Class', 'Sample'])
split_2 = tr['Class']
split_1.head()
```

```python
split_2.head()
```

```python
#Splitting the data for training and testing 
split_1_train, split_1_test, split_2_train, split_2_test = train_test_split(split_1,split_2, test_size=0.3, random_state = 4, shuffle = True) 
```

```python
#defining and fitting the decision tree model
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(split_1_train, split_2_train)
```

```python
#prediction on testing data
results_predicted = decision_tree_model.predict(split_1_test)
```

```python
#confusion matrix
print(f"confusion_matrix is\n{confusion_matrix(split_2_test,results_predicted)}")
```

```python
classified_report_results = classification_report(split_2_test, results_predicted)
print("\n classification report:")
print(classified_report_results)
```

```python
#Accuracy
accuracy_results= (accuracy_score(split_2_test, results_predicted)*100)
print("accuracy:",accuracy_results)
```

```python
error_rate= 1 - accuracy_score(split_2_test,results_predicted)
print(f"error rate is {error_rate:.2f}")
```

```python
#Plotting the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize = (50,30), dpi = 250)
plot_tree(decision_tree_model, fontsize = 20, filled = True,
         feature_names = split_1.columns.tolist())
```
