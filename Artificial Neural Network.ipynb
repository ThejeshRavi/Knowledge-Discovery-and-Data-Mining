```python
#Importing the libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Loading the dataset
df = pd.read_csv(r'E:\wisc_bc_ContinuousVar.csv')
df.head()
```

```python
# Label encoding 
label_enc = LabelEncoder()
df['diagnosis'] = label_enc.fit_transform(df['diagnosis'])
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Data Standardizing
scaler_results = StandardScaler()
X_train = scaler_results.fit_transform(X_train)
X_test = scaler_results.transform(X_test)
```

```python
model_ANN = Sequential()
model_ANN.add(Dense(5, input_dim=X_train.shape[1], activation='relu'))
model_ANN.add(Dense(1, activation='sigmoid'))

# Compiling the model
model_ANN.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_ANN.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)
```

```python
# Predictions on the test set
y_pred_prob = model_ANN.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)
```

```python
# Evaluate the model
accuracy_results = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy_results}")
```

```python
cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{cm}")
```

```python
cr_results = classification_report(y_test, y_pred)
print(f"Classification Report:\n{cr_results}")
```
