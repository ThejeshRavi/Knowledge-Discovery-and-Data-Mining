```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data
tr =pd.read_csv(r"E:\breast-cancer-wisconsin.csv")
tr =pd.DataFrame(tr)
tr.head()
```

```python
# Summarizing each column
getting_summary = tr.describe()
print(getting_summary)
```

```python
#identifying the missing values
tr = tr.replace("?", pd.NA)
tr = tr.apply(pd.to_numeric, errors = 'ignore')
getting_missing_values =tr.isnull().sum()
print(getting_missing_values)
```

```python
#Replacing the missing values with column mean
average_F6= (tr['F6'].mean())
print(f"Mean of the column F6:{average_F6}")
tr['F6'].fillna(average_F6, inplace = True)
```

```python
#Displaying the frequency table of 'class' vs F6
frequency_display = pd.crosstab(tr['Class'],tr['F6'])
print(f'frequency_display:\n{frequency_display}')
```
